# Seaborn
## Grids

```python
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
%matplotlib inline
```

### pairgrid
It is like pairplor bit in this case we can choose direclty how to show the data

```python
iris = sb.load_dataset('iris')
iris.head()
```

```python
g = sb.PairGrid(iris)
#g.map_diag(sb.distplot)
g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(sb.kdeplot)
```

### Facet Grid
FacetGrid is the general way to create grids of plots based off of a feature:

```python
tips = sb.load_dataset('tips')
tips.head()
```

```python
g = sb.FacetGrid(tips, col="time", row="smoker")
```

```python
g = sb.FacetGrid(tips, col="time",  row="smoker")
#g = g.map(plt.scatter, "total_bill").add_legend()
#g = g.map(sb.distplot, "total_bill").add_legend()
g = g.map(plt.hist, "total_bill").add_legend()
```

### JointGrid
JointGrid is the general version for jointplot() type grids, for a quick example:

```python
g = sb.JointGrid(x="total_bill", y="tip", data=tips)
g = g.plot(sb.regplot, sb.distplot)
```
